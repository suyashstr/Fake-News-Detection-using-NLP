```python
#importing Dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("train.csv")
```

```python
# Handling missing values by dropping rows where 'text' is NaN
data_cleaned = data.dropna(subset=['text'])
```

```python
# Visualizing the distribution of real and fake news with a count plot
plt.figure(figsize=(7, 7))
sns.set(style="darkgrid")
color = sns.color_palette("Set1")
ax = sns.countplot(x="label", data=data_cleaned, palette=color)
ax.set(xticklabels=['FAKE', 'REAL'])
plt.title("Data distribution of fake and real news")
plt.show()
```

```python
# Pie chart for fake and real news distribution
plt.figure(figsize=(7, 7))
labels = ['Fake News', 'Real News']
sizes = data_cleaned['label'].value_counts()
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # explode 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Pie chart of Fake and Real News Distribution")
plt.show()
```

```python
# Visualizing the top 10 authors with the most articles
plt.figure(figsize=(10, 6))
top_authors = data_cleaned['author'].value_counts().nlargest(10)
sns.barplot(x=top_authors.values, y=top_authors.index, palette="Blues_r")
plt.title("Top 10 Authors with the Most Articles")
plt.xlabel("Number of Articles")
plt.ylabel("Author")
plt.show()
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_cleaned['text'], data_cleaned['label'], test_size=0.2, random_state=42)
```

```python
# Initializing TfidfVectorizer with English stop words and a max document frequency of 0.7
tfidf_vect = TfidfVectorizer(stop_words="english", max_df=0.7)
tfidf_train = tfidf_vect.fit_transform(X_train)
tfidf_test = tfidf_vect.transform(X_test)
```

```python
# Initializing PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(tfidf_train, y_train)
```

```python
# Predicting on the test set
y_pred = pac.predict(tfidf_test)
```

```python
plt.figure(figsize=(7, 7))
sns.countplot(x=y_pred, palette="Set2")
plt.title("Predicted Distribution in the Test Set (Fake vs Real)")
plt.xlabel("Label (0: Fake, 1: Real)")
plt.show()
```

```python
# Pie chart for predicted distribution
plt.figure(figsize=(7, 7))
labels = ['Fake News', 'Real News']
sizes = [sum(y_pred == 0), sum(y_pred == 1)]
colors = ['orange', 'lightgreen']
explode = (0.1, 0)  # explode 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title("Pie chart of Predicted Fake and Real News Distribution")
plt.show()
```

```python
# Calculating accuracy
accuracy = round(accuracy_score(y_pred, y_test) * 100, 2)
print(f"Accuracy Score: {accuracy}%")
```
